```python
#This cell check the combinations of countries mentioned in the annoucements

# Handle missing or non-string values in 'country' column
def normalize_countries(country):
    if pd.isna(country) or not isinstance(country, str):
        return country  # Keep NaN or non-string values as they are
    return ', '.join(sorted(country.split(', ')))

df['sorted_countries'] = df['country'].apply(normalize_countries)

# Filter out single-country entries and count combinations
multi_country_combinations = df['sorted_countries'].dropna().apply(lambda x: x if len(x.split(', ')) > 1 else None)
country_combinations_count = multi_country_combinations.value_counts()

# Convert to DataFrame for better formatting and display
country_combinations_df = country_combinations_count.reset_index()
country_combinations_df.columns = ['Country Combination', 'Frequency']

# Sort by Frequency
country_combinations_df = country_combinations_df.sort_values(by='Frequency', ascending=False)

print(country_combinations_df[:20])
```
