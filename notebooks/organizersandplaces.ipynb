```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the database to a dataframe
df = pd.read_csv('../data/lanternworlddata.tsv', sep='\t')
```

```python
#create bar charts of the top-five organizers in Brussels and Antwerp in the three periods

# Define the periods and cities of interest
periods = ['1902-1904', '1914-1918', '1922-1924']
cities = ['Brussel', 'Antwerpen']

# Create a 2x3 subplot structure (2 rows and 3 columns)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(30, 10))

# Iterate over each city and period
for row, city in enumerate(cities):
    for col, period in enumerate(periods):
        # Filter the data for the current city and period
        period_df = df[(df['STAD STANDAARD_GEGROEPEERD'] == city) & (df['period'] == period)]
        total_occurrences = len(period_df)
        top_5_organizers = period_df['organisator1_gestandaardiseerd'].value_counts().head(5) / total_occurrences * 100
        
        # Plot on the subplot with black bars
        top_5_organizers.plot(kind='barh', color='black', ax=axes[row, col])
        
        # Set the x-label and title
        axes[row, col].set_xlabel('Percentage of Occurrences')
        axes[row, col].set_title(f'Top 5 Organizers in {city} for {period}')
        
        # Remove the y-labels
        axes[row, col].set_ylabel('')
        
        # Invert the y-axis
        axes[row, col].invert_yaxis()

# Adjust the layout
plt.tight_layout()

# Adjust the vertical spacing between rows of plots
plt.subplots_adjust(hspace=0.5)  # Increase this value as needed

#The code below saves the Figure X of the article
plt.savefig('top_organizers_brussel_antwerpen.png', dpi=300)

# Display the plot
plt.show()
```
